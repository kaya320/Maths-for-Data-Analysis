# Классическое определение вероятности 

Вероятность - число, характеризующее степень возможности появления события.
Вероятностью события A называют отношение числа благоприятствующих исходов к общему числу всех исходов

Событие называют достоверным, если в результате испытания оно обязательно происходит

**P(A) = n/N**

# Сложение вероятностей 

**Несовместные события** - если в одном и том же испытании они не могут произойти одновременно.

Если событие C - наступление одного из двух несовместных событий A или B, то вероятность события C - сумма вероятностей событий A и B:

**p(C)=p(A)+p(B)**

**Совместные события**

Вероятность суммы двух совместных событий равна сумме вероятностей этих событий минус вероятность их произведения

**p(A+B) = p(A)+p(B) - p(A⋅B)**

# Произведение вероятностей

**Независимые события** - если наступление одного из них не влияет на наступление другого.

Если событие C означает совместное наступление двух независимых событий A и B, то вероятность такого события C равна произведению вероятностей событий A и B 

**p(C)=p(A)⋅p(B)**
